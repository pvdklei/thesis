```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv("results_prepped.csv")
```

```python
data4 = data[data["p"] == 0.4]
data8 = data[data["p"] == 0.8]
```

```python
fig, ax = plt.subplots(figsize=(400, 400))
ax = sb.barplot(ax=ax, data=data, y="id", x="average_drawtime", hue="p")
ax.set(ylabel=None)
ax.set(xlabel="Average Drawtime (ms)")
ax.legend(title="Vertices per fragment")
"Plot"
```

```python
data4["average_drawtime"].plot.barh(xlim=(6, 8), ylabel="Draw Time (ms)")
```

```python
data8["average_drawtime"].plot.barh(xlim=(7.50, 8.7), ylabel="Draw Time (ms)")
```

```python
data.groupby("p")["average_drawtime"].plot.barh(color="green yellow blue red".split())
plt.show()
```
